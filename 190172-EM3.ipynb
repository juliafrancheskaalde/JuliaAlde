```python
# write code here
#create a function that cleans the birthmonth column
months = {"1":"January",'2':"February",'3':"March",'4':"April",'5':"May",'6':"June",'7':"July",'8':"August",'9':"September",'10':"October",'11':"November",'12':"December",
         "Jan":"January",'Feb':'February','Mar':'March','Apr':'April','Jun':'June','Jul':'July','Aug':'August','Sep':'September','Oct':'October','Nov':'November','Dec':'December'}

def clean_month(date):
    if date in months.keys():
        return months[date]
    else:
        return date
```

```python
#apply function to birthmonth column
recweek['birthmonth']= recweek['birthmonth'].apply(clean_month)
#dump
recweek
```

```python
# Clean Year
# dump recweek data
recweek
```

```python
#make a function that cleans the birthyear column
from datetime import datetime
def clean_year(year):
    year= str(year)
    year= year.lstrip()
    if len(year)== 2:
        date= datetime.strptime(year,"%y")
        new_year= date.strftime("%Y")
        return new_year
    else:
        return year
```

```python
#apply function to birthyear column
recweek['birthyear']= recweek['birthyear'].apply(clean_year)
#dump
recweek
```

```python
# Clean Course
# dump recweek data
recweek
```

```python
#dictionary of all possible inputs for course format
courses= {'ITE':['Information Technology Entrepreneurship','ITE','BS ITE','BS Information Technology Entrepreneurship'],
        'ME':['Management Engineering','ME','BS ME','BS Management Engineering'],
        'CTM':['Communications Technology','BS CTM','CTM','BS Communications Technology'],
        'LM':['Legal Management','BS LM','LM','BS Legal Management'],
        'MGT':['Management','BS MGT','MGT','BS Management']}

#make a function to clean the course column
def clean_course(course):
    if course in courses['ME']:
        return "BS Management Engineering"
    elif course in courses['CTM']:
        return "BS Communications Technology"
    elif course in courses['LM']:
        return "BS Legal Management"
    elif course in courses['MGT']:
        return "BS Management"
    else:
        return "BS IT Entrepreneurship"
```

```python
#apply to course column
recweek['course']= recweek['course'].apply(clean_course)
#dump
recweek
```

```python
# write code here

#dictionary of prefixes of the different networks
networks={'GLOBE_TM':['0917','0925','0905','0906','0915','0916','0926','0927','0935','0936','0937','0945','0953','0954','0955','0956','0965','0966','0967','0975','0977','0978','0979','0995','0996','0997'], 
        'SMART':['0908','0918','0919','0920','0921','0928','0929','0939','0946','0947','0949','0951','0961','0998','0999']}

#function that returns the network, given the prefix
def carrier(number):
    number= str(number)
    prefix = number[0:4]
    if prefix in networks['GLOBE_TM']:
        return "Globe"
    elif prefix in networks['SMART']:
        return "Smart"
```

```python
#add a column for the carrier/network
final_recweek["carrier"]= final_recweek['phonenumber'].apply(carrier)
#dump
final_recweek
```

```python
#count the occurances of the each carrier/network in the column 'carrier'
count_carrier=final_recweek.groupby('carrier').count()['firstname']
#convert to dataframe
count_carrier.to_frame(name='count')
```
